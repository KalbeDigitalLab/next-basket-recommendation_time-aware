```python
import pandas as pd
from datetime import datetime, timezone
```

```python
df = pd.read_csv('data/EMOS.csv')
df
```

```python
df['invoice_date'] = df['invoice_date'].apply(lambda x: int(datetime.strptime(x[:10], '%Y-%m-%d').replace(tzinfo=timezone.utc).timestamp()))
df
```

```python
df.sort_values(by=['ship_to_id_emos', 'invoice_date']).to_csv('data/EMOS-prep.txt', sep='\t', index=False, header=False)
```
